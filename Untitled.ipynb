```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from crtbp_prop import propCrtbp, propNRevsPlanes

# расчет точек либрации:
from lagrange_pts import lagrange_pts
from find_vel import findVPlanes
```

```python
# Constants
Sm = 1.9891e30  # mass of the Sun
Em = 5.97e24  # mass of the Earth
ER = 1.496e8  # Sun-Earth distance
T = 365*24*60*60  # Период обращения земли вокруг солнца
# km ~ 1/ER
# sec ~ 2pi/T
# km/s ~ T/(2pi*ER)
mu1 = Sm/(Sm + Em)  # CRTBP main coefficient

vcoef = T/(2*np.pi*ER)
earth = np.array([mu1,0,0])
L = lagrange_pts(mu1)
L2 = L[1, 0]

x0km = -100000
z0km = 0
v0kms = 0.1

# начальный вектор
s0 = np.array([L2+x0km/ER, 0., z0km/ER, 0, v0kms*vcoef, 0])
```

```python
xl = earth[0] + 100000/ER
xr = L[1,0] - 100000/ER
#xr = L[1,0]+(L[1,0]-earth[0])
zl = -1.*(xr-xl)
zr = xr-xl
```

```python
xs = np.linspace(xr,xl,100)
zs = np.linspace(zr,zl,100)
```

```python
X,Y = np.meshgrid(xs,zs)
```

```python
plt.figure(figsize=(20,20))
plt.plot(X,Y,'.b');
plt.plot(earth[0],earth[1],'or')
plt.plot(L[1,0], L[1,1], '*r',markersize=20)
plt.tight_layout()
```

```python
from IPython.display import clear_output
```

```python
rez = []
n1 = 0
for z_j in zs:
    xl_prev = earth[0]+100000/ER
    xr_prev = L2+(L2-earth[0])
    for x_i in xs:
        n1+=1
        planes = [xl_prev-60000/ER,xr_prev+60000/ER,1]
        s0 = [x_i,0,z_j,0,v0kms*vcoef,0]
        out = propNRevsPlanes(mu1,s0,0,planes,20);
        if type(out) == dict:
            print('{}/10000'.format(n1))
            clear_output(wait=True)
            rez.append({'x':x_i,'z':z_j,'conv':0,'arr':[],'dv':out['v']})
            break
        else:
            print('{}/10000'.format(n1))
            clear_output(wait=True)
            rez.append({'x':x_i,'z':z_j,'conv':1,'arr':out,'dv':[]})
            xl_prev = out[:,0].min()
            xr_prev = out[:,0].max()
```
